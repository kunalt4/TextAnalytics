```python
import nltk
import string
from nltk.corpus import stopwords
```

```python
documents = {}
tokens = []
pos_tokens = []
```

```python
for doc in range(1,11):
    with open('doc' + str(doc) + '.txt') as f:
        documents['doc' + str(doc)] = f.read()
```

```python
for key, doc in documents.items():
    documents[key] = doc.translate(doc.maketrans('', '', string.punctuation))
    tokens.append(nltk.word_tokenize(documents[key]))
```

```python
tokens
```

```python
stop = set(stopwords.words('english'))
for docs in range(len(tokens)):
    tokens[docs] = [w.lower() for w in tokens[docs]]
    tokens[docs] = [w for w in tokens[docs] if w not in stop]
    pos_tokens.append(nltk.pos_tag(tokens[docs]))
```

```python
pos_tokens
```

```python
def convert_tags_to_wordnet(tag):
    if tag.startswith('V'):
        return 'v'
    elif tag.startswith('J'):
        return 'a'
    elif tag.startswith('R'):
        return 'r'
    elif tag.startswith('N'):
        return 'n'
    else:
        return None
```

```python
lemma = nltk.WordNetLemmatizer()
lemmatized_words = []
```

```python
for token in pos_tokens:
    w = []
    for word in token:
        converted_tag = convert_tags_to_wordnet(word[1])
        if(converted_tag is not None):
            out = lemma.lemmatize(word[0],converted_tag)
            w.append(out)
    lemmatized_words.append(w)
lemmatized_words
```

```python
seperator = ' '
for words in range(len(lemmatized_words)):
    lemmatized_words[words] = seperator.join(lemmatized_words[words])
lemmatized_words
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
vectorizer = CountVectorizer()
```

```python
word_count_vector = vectorizer.fit_transform(lemmatized_words)
```

```python
print(word_count_vector.shape)
```

```python
print(word_count_vector.toarray())
```

```python
print(vectorizer.vocabulary_)
```

```python
import pandas as pd
tf = pd.DataFrame(word_count_vector.todense(), columns=vectorizer.get_feature_names(), index=documents.keys())
```

```python
from nltk.collocations import *
from nltk.tokenize import word_tokenize
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
lemmatized_words
```

```python
wc_text = " ".join(review for review in lemmatized_words)
wc_text
finder = BigramCollocationFinder.from_words(word_tokenize(wc_text))
for i in finder.score_ngrams(bigram_measures.pmi):
    print(i)
```

```python
finder = BigramCollocationFinder.from_words(word_tokenize(wc_text))
finder.apply_freq_filter(2)
for i in finder.score_ngrams(bigram_measures.pmi):
    print(i)
```

```python
f = open('spam_tweets.txt',encoding='utf-8')
spam_tweets = f.read()
f = open('random_tweets.txt',encoding='utf-8')
random_tweets = f.read()
spam_tweets
```

```python
random_tweets
```

```python
spam_tweets = spam_tweets.translate(spam_tweets.maketrans('', '', string.punctuation))
spam_tweets = [w for w in word_tokenize(spam_tweets) if w not in stop]
spam_tweets = " ".join(word for word in spam_tweets)
spam_tweets
```

```python
random_tweets = random_tweets.translate(random_tweets.maketrans('', '', string.punctuation))
random_tweets = [w for w in word_tokenize(random_tweets) if w not in stop]
random_tweets = " ".join(word for word in random_tweets)
random_tweets
```

```python
import math
def entropy(labels):
    freqdist = nltk.FreqDist(labels)
    probs = [freqdist.freq(l) for l in freqdist]
    return -sum(p * math.log(p,2) for p in probs)
```

```python
print(entropy(word_tokenize(spam_tweets)))
```

```python
print(entropy(word_tokenize(random_tweets)))
```

```python
combined_tweets = spam_tweets + random_tweets
```

```python
print(entropy(word_tokenize(combined_tweets)))
```
